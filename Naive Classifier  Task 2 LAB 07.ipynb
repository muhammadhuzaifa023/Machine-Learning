```python
#Splitting the Dataset
#Taking Job scheduling  Dataset
dependent1=df["family history"]
dependent1.tail()
```

```python
# independent variables 
independent1= df.drop('family history',axis='columns')
independent1
```

```python
feature=[feature for feature in df.columns if df[feature].dtypes == 'O']
feature
```

```python
Independent1=independent1.drop(' Age (linear)',axis='columns')
Independent1
```

```python
#Divide the dataset into traintestsplit
#trainData=60%
#testData=40%
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(Independent1,dependent1,test_size=0.4)
```

```python
from sklearn.naive_bayes import GaussianNB
model=GaussianNB()
model.fit(X_train,y_train)
```

```python
print("Training Accuracy= {}".format(model.score(X_train,y_train)))
```

```python
print("Testing Accuracy= {}".format(model.score(X_test,y_test)))
```

```python
pred1 = model.predict(X_test)
print(pred1)
```
