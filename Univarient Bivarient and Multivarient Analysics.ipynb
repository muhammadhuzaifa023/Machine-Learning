# Univarient,Bivarient and Multivarient Analysics


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sn
```

```python
dataset=pd.read_csv('iris.csv')
dataset.head(56)
```

```python
dataset.columns
```

```python
dataset.shape
```

###Univarient Analysics
dataset.loc[dataset['species']=='Setosa']



```python
 df_setsose=dataset.loc[dataset['variety']=='Setosa']
df_setsose
```

```python
 df_Virginica=dataset.loc[dataset['variety']=='Virginica']
df_Virginica
```

```python
 df_Versicolor=dataset.loc[dataset['variety']=='Versicolor']
df_Versicolor
```

```python
plt.plot(df_setsose['sepal.length'],np.zeros_like(df_setsose['sepal.length']),'o')
plt.plot(df_Virginica['sepal.length'],np.zeros_like(df_Virginica['sepal.length']),'o')
plt.plot(df_Versicolor['sepal.length'],np.zeros_like(df_Versicolor['sepal.length']),'o')
plt.xlabel('Petal_Length')
plt.show()
```

```python
###Bivarite Analysics
sn.FacetGrid(dataset,hue='variety',size=5).map(plt.scatter,"sepal.length","sepal.width").add_legend();
plt.show()
```

```python
##Multivarient Analysics
sn.pairplot(dataset,hue='variety',size=3)
```
